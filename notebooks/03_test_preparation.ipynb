```python
import pandas as pd
import os
import shutil

data_folder_path = "../datasets/bukva/trimmed/"
data_csv_path = "../datasets/bukva/annotations.csv"
test_folder_path = "../datasets/bukva_processed/test"
test_csv_path = "../datasets/bukva_processed/test_annotations.csv"

os.makedirs(test_folder_path, exist_ok=True)

# Load the original csv file
df = pd.read_csv(data_csv_path, sep="\t")

df.head()
```

```python
# Filter test data
test_df = df[df['train'] == False].copy()
test_df.head()
```

```python
for index, row in test_df.iterrows():
    file_name = f"{row['attachment_id']}.mp4"

    src = os.path.join(data_folder_path, file_name)
    dst = os.path.join(test_folder_path, file_name)

    if os.path.exists(src):
        shutil.move(src, dst)
    else:
        print(f"Warning: File {file_name} not found in {data_folder_path}")

# Drop the redundant columns
test_df = test_df.drop(columns=['user_id', 'train'])

test_df.to_csv(test_csv_path, sep="\t", index=False)
print(f"Successfully created {test_csv_path} with {len(test_df)} entries.")
```
